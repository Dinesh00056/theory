```python
# What is Naive Bayes......
# Naive bayes is a probabilistic machine learning model based on bayes theorem it is mainly used for clasification problem suuch as image spam detetction medical 
# diagonsis it is called naive because it assumes all features are independent to each other and it is called bayes because it is based on bayes theorem of 
# probability. Bayes Therorem is used to update the probability of a hypothesis class based on new evidence 

# formula=probability of c/x=propability of x/c into probalitity of c / x
# where probability of c/x it is a probability of class c given feature x
# probability of x/c= probability of features x in given class c
# probability of c= prior probability of class 
# probability of x= (evidence) probability of feature of all classes 
```

# Advantages
It is very fast and efficient
It works with high dimensional data
It requires small training data 
It handles noisy data

```python
# Limitation
# 1) It assumes independent of feature 
#2) It is not suitable for features with strong correlation
```

```python
# Suppose we want to classify whether a person buy a computer(yes/no) based on two features 
# 1) Age=young middle old
# 2) Income= High, medium,low

# Person     Age        Incomes          Buy_Computer
# 1          Young       High             No
# 2          Young       Medium           No
# 3          Middle      High             Yes
# 4          Old         Medium           Yes
# 5          Old         Low              Yes
# 6          Middle      Low              No

# step on 1 --> We calculate how many people bought verses how many people didnot buy a computer 
# total=6
# buy=3/Yes
# Not Purchased=3/NO
# probability of Yes=3/6=0.5
# probability of NO=3/6=0.5

# Step 2
# Now we compute probability of features given the (age) class 
# for class=yes
# probability of age middle=1/3
# probability of age old=2/3
# probability of high income=1/3

# For Class =No
# probability of age young=2/3
# probability of age middle=1/3
# probability of age old=0/3

# Step 3 
# now suppose we want to predict for a new person whose 
# age=old income =medium
# probability of yes/old and medium= prob of yes into prob old / yes into prob medium/yes
# probability of yes/old and medium=0.5*2/3*1/3=0.11111

# probability of NO/old and medium

# Step 4 
# Probability of yes/old and medium is =0.111
# Probability of No/old and medium is = 0

# the model will predict the yes as the probability of yes is high
```
